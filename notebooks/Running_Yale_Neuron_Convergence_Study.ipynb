```python
import yaleneuronsim as yn
import matplotlib.pyplot as plt
```

```python
filename='../cells/0-2a.CNG.swc'
cell = yn.yaleneuronsim(filename)
```

```python
pw=list(range(8,-4,-1))
dx=[pow(2,j) for j in pw]

print('Running with dx = ...',end='')
for delta in dx:
    print(str(delta),' ',end='')
    cell.set_discretization(delta)
    nlocs=cell.get_swc_locs(False)

    # set some parameters
    v_initial=-65; end_time=100
    dur=100; delay=0.0; amp=0.9
    vdata,t=cell.setup_run(v_initial,end_time,dur,amp,delay)

    k = list(vdata.keys())
    
    plt.plot(t,vdata[k[0]],linewidth=0.65,label='dx='+str(delta)+'um')
    
plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
plt.show()
```
